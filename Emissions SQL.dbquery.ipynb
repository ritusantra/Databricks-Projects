```sql
-- 1
select * from emissions_data;

-- 2
select latitude, longitude, 
`GHG emissions mtons CO2e` as emissions
from emissions_data;

-- 3

select county_state_name, population,
  cast(
    replace(
      `GHG emissions mtons CO2e`, 
      ',', 
      ''
    ) as double
  ) / cast(
    replace(
      population, 
      ',', 
      ''
    ) as double
  ) as emissions_per_person
from emissions_data
order by emissions_per_person desc
limit 10;

-- 4.

select state_abbr, 
  sum(cast(
    replace(
      `GHG emissions mtons CO2e`, 
      ',', 
      ''
    ) as double
  )) as total_emissions
from emissions_data
group by state_abbr
order by total_emissions desc;

-- 5. 

with state_emissions as (
  select state_abbr, 
    sum(cast(replace(`GHG emissions mtons CO2e`, ',', '') as double)) as total_emissions
  from emissions_data
  group by state_abbr
),
top10 as (
  select state_abbr, total_emissions
  from state_emissions
  order by total_emissions desc
  limit 10
)
select 
  sum(top10.total_emissions) / (select sum(total_emissions) from state_emissions) * 100 as top10_pct
from top10;

-- 6.

select county_state_name, population,
  cast(
    replace(
      `GHG emissions mtons CO2e`, 
      ',', 
      ''
    ) as double
  ) as total_emissions
from emissions_data
order by total_emissions desc
limit 10;
```
