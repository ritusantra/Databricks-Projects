```python
%sql

-- check duplicate cst_id
select  
cst_id, count(*)
from sales_catalog.silver.s_crm_cust_info
group by cst_id 
having count(*)>1 or cst_id is null
```

```python
%sql

-- Check for unwanted spaces
select * 
from sales_catalog.silver.s_crm_cust_info
where cst_firstname != trim(cst_firstname);


select * 
from sales_catalog.silver.s_crm_cust_info
where cst_lastname != trim(cst_lastname)
```
