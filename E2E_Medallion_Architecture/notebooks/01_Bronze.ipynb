```python
print('Creating Bronze Schema...')

spark.sql('''
    CREATE SCHEMA IF NOT EXISTS sales_catalog.bronze
''')

print('Bronze Schema Creation Successful!')
```

```python
print('Creating Bronze Tables...')

spark.sql('''
    CREATE OR REPLACE TABLE sales_catalog.bronze.b_crm_cust_info
    (
        cst_id             INT,
        cst_key            STRING,
        cst_firstname      STRING,
        cst_lastname       STRING,
        cst_marital_status STRING,
        cst_gndr           STRING,
        cst_create_date    DATE
    )
''')
```

```python
spark.sql('''
    CREATE OR REPLACE TABLE sales_catalog.bronze.b_crm_prd_info
    (
        prd_id       INT,
        prd_key      STRING,
        prd_nm       STRING,
        prd_cost     DOUBLE,
        prd_line     STRING,
        prd_start_dt DATE,
        prd_end_dt   DATE
    )
''')
```

```python
spark.sql('''
    CREATE OR REPLACE TABLE sales_catalog.bronze.b_sales_details
    (
        sls_ord_num  STRING,
        sls_prd_key  STRING,
        sls_cust_id  INT,
        sls_order_dt INT,
        sls_ship_dt  INT,
        sls_due_dt   INT,
        sls_sales    DOUBLE,
        sls_quantity INT,
        sls_price    DOUBLE
    )
''')
```

```python
spark.sql('''
    CREATE OR REPLACE TABLE sales_catalog.bronze.b_cust_az12
    (
        CID   STRING,
        BDATE DATE,
        GEN   STRING
    )
''')
```

```python
spark.sql('''
    CREATE OR REPLACE TABLE sales_catalog.bronze.b_loc_a101_cntry
    (
        CID   STRING,
        CNTRY STRING
    )
''')
```

```python
spark.sql('''
    CREATE OR REPLACE TABLE sales_catalog.bronze.b_px_cat_g1v2
    (
        ID          STRING,
        CAT         STRING,
        SUBCAT      STRING,
        MAINTENANCE STRING
    )
''')

print('Bronze Tables Creation Successful!')
```
